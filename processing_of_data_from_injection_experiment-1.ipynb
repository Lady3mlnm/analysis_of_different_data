#### Displaying of raw experimental data

```python
import matplotlib.pyplot as plt
from collections import namedtuple
```

```python
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['font.size'] = 12
```

```python
# source data
Graph = namedtuple('Graph', 'data, label, color, linestyle, linewidth, marker, markersize, alpha, zorder',
                   defaults=(None, 'grey', '-', 1, '.', 13, 1, None))

experiment1_day1 = Graph(data={1:1162, 4:1624},
                    color='lightseagreen',
                    linestyle=':',
                    zorder=2)

experiment1 = Graph(data={4:1624, 8:908, 11:567, 14:434},
                    label='intramuscular injection',
                    color='lightseagreen',
                    zorder=3)

experiment2 = Graph(data={2:1555, 6:1167, 10:556, 14:252},
                    label='subcutaneous injection',
                    color='blue',
                    zorder=5)

experiment2_check = Graph(data={14: 280},
                    label='subcutaneous injection - verification',
                    color='deepskyblue',
                    linewidth=0,
                    zorder=4)

bound = Graph(data={-1: 495, 100: 495},
              label='lower bound of desired level',
              color='#BFBF00',
              linewidth=1.4,
              alpha=0.5,
              marker=None,
              zorder=1)
```

```python
def create_graph(*series, shift_x=-0.6, nmb_displayd_days=14, title='Raw measurement results'):
    for ser in series:
        values_x = tuple(map(lambda x: x+shift_x, ser.data.keys()))
        values_y = tuple(ser.data.values())

        plt.plot(values_x, values_y,
                 label=ser.label,
                 color=ser.color,
                 linestyle=ser.linestyle,
                 linewidth=ser.linewidth,
                 marker=ser.marker,
                 markersize=ser.markersize,
                 alpha=ser.alpha,
                 zorder=ser.zorder
                 )
    
    # configure the chart
    plt.title(title)
    plt.xticks(range(nmb_displayd_days+1))
    plt.legend()
    plt.grid(color='wheat', linewidth=0.25)
    plt.xlim(0, nmb_displayd_days)
    plt.ylim(0)
    plt.show()


create_graph(experiment1_day1, experiment1, experiment2, experiment2_check, bound)
```
